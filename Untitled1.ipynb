```python
import pySibernetic
```

```python
config=pySibernetic.owConfigProperty('worm')
```

```python
config.TimeLim = 10.
config.LogStep = 20 #every 1e-4 sec.
config.TimeStep = 5e-6

sim=pySibernetic.owPhysicsFluidSimulator(config)
```

```python
m=pySibernetic.MuscleSimulation()
m.increment=5;
```

```python
import matplotlib.animation
```

```python
for i in range(25): 
        sim.simulationStep(True)
        m_t=m.run()
        sim.MuscleAtcivationSignal[:96]=m_t
        sim.pushMuscleActivity2openCLsolver()
    
```

```python
close()
f=figure()
ax=subplot(111)
```

```python
def anim_worm(t):
    for i in range(25): 
        sim.simulationStep(True)
        m_t=m.run()
        sim.MuscleAtcivationSignal[:96]=m_t
        sim.pushMuscleActivity2openCLsolver()
    r=sim.Position.reshape((-1,4))
    r1 = r[(r[:,3]>1.75) * (r[:,3]<2.55)  ,:3]
    ax.clear()
    s=ax.scatter(r1[0::1,2],r1[0::1,0],s=1)
    ax.set_aspect(1)
    ax.set_title("t={}".format(t))
    xlim(0,700)
    ylim(0,100)
```

```python
matplotlib.animation.FuncAnimation(f,anim_worm,500)
```
